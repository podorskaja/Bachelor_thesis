```python
from sklearn.cross_decomposition import CCA
```

```python
import scanpy.api as sc
import pandas as pd
import loompy
import numpy as np
import anndata as ad
import matplotlib.pyplot as plt
```

```python
data1_path = './pairwise/t4k.csv'
data2_path = './pairwise/pbmc4k.csv'
```

```python
X = pd.read_csv(data1_path, sep=",", header=None)
Y = pd.read_csv(data2_path, sep=",", header=None)
```

```python
cca = CCA(n_components=1)
```

```python
%%time
cca.fit(X, Y)
```

```python
X_c, Y_c = cca.transform(X, Y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_c, Y_c, 
            marker="*", c="b")
```
